```python
new_df = df[df['area'].notnull()]
new_df.isnull().sum()
```

```python
new_df.head(7)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
lablelencoderlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for i in lablelencoderlist:
    new_df[i] = enc.fit_transform(new_df[i])

new_df.head(7)
```

```python
new_df['furnishingstatus'] = new_df['furnishingstatus'].map({'furnished':2,'unfurnished':0, 'semi-furnished':1})
new_df.head(15)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
featurescal = ['area','bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']
new_df[featurescal] = scaler.fit_transform(new_df[featurescal])
new_df
```

```python
temp = new_df.corr()
print(temp['price'].sort_values(ascending=False))
```

```python
# basement and hotwaterheating lowest
new_df = new_df.drop(['basement','hotwaterheating'],axis=1)
new_df.head()
```

```python
# linear regression

X = new_df.drop('price', axis=1)
y = new_df['price']

from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(xTrain, yTrain)
yPrediction = clf.predict(xTest)

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(yTest, yPrediction)

import numpy as np
rmse = np.sqrt(mse)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
# logistic reg

import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# thresholds = [3000000, 10000000]
thresholds = [new_df['price'].median()]
y_train = pd.cut(yTrain, bins=[-float('inf')] + thresholds + [float('inf')], labels=False)
y_test = pd.cut(yTest, bins=[-float('inf')] + thresholds + [float('inf')], labels=False)

model = LogisticRegression()
model.fit(xTrain, y_train)

predictions = model.predict(xTest)


print(f"Accuracy: {accuracy_score(y_test, predictions)}")
```
